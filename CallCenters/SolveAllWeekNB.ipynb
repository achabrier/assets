```python
from dd_scenario import *

# In order to use the solve() function, you must provide an API key when creating the client
client = Client(pc=pc, apikey=apikey)
decision = client.get_model_builder(name="CallCentersStaffPlanningExperiment")
scenario = decision.get_scenario(name="Baseline")
```

```python
sc_name = "AllWeeks"
copy = decision.get_scenario(name=sc_name)
if (copy != None):
    print(" Deleting old...")
    decision.delete_container(copy)
scenario = scenario.copy(sc_name)
```

```python
N_WEEKS = 2
```

```python
import pandas as pd

finalSchedulesSol = pd.DataFrame()
finalCouverture = pd.DataFrame()

for week in range(N_WEEKS):
    print(" Solving week ",week)
    parameters = scenario.get_table_data('parameters')
    parameters = parameters.set_index(['name'])
    parameters.value['handleWeek'] = week
    parameters = parameters.reset_index()
    print(parameters)
    scenario.add_table_data("parameters", parameters, category='input')
    print(" Solving...")
    scenario.solve(display_log=True, log_lines=None)

    finalSchedulesSol = finalSchedulesSol.append(scenario.get_table_data("weekSchedulesSol"))
    print(finalSchedulesSol.size)
    finalCouverture = finalCouverture.append(scenario.get_table_data("weekCouverture"))
    print(finalCouverture.size)
        
    
scenario.add_table_data("finalSchedulesSol", finalSchedulesSol, category="output")
scenario.add_table_data("finalCouverture", finalCouverture, category="output")
print("Done!")
```
